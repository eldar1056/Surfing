```python
def c_0(x):
    return math.sin(math.pi*x)
#     if 1 < x < 2:
#          return 1
#     else:
#          return 0
```

```python
L = 10 # весь интервал координаты
TAU = 0.01 # разбиение по времени
H = 0.01 # разбиение по координате
u = 0.1 # константа скорости
T = 5 # время для которого требуется найти распределение c
A = u*TAU/H # коэффициент пропорциональности в форуле
```

```python
def next_layer(layer, t):
    output = [] #[c_boundary(t)]
    for i in range(len(layer)):
        #if i > 0:
        output.append(layer[i]-A*(layer[i]-layer[i-1]))
    return output
```

```python
x_range = [alpha*H for alpha in range(math.ceil(int(L/H)+1))]

current_layer = [c_0(x) for x in x_range]

for step in range(1, int(T/TAU)+1):
    current_layer = next_layer(current_layer, step*TAU)

plt.plot(x_range, [c_0(x) for x in x_range])
plt.plot(x_range, current_layer)
```
